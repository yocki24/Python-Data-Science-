```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# code here
fig, ax = plt.subplots(figsize=(18,8),nrows=1, ncols=3, sharey=True)
ax[0].plot(x, tan, color='b', label='tanx')
ax[1].plot(x, sin, color='r', label='sinx')
ax[2].plot(x, cos, color='g', label='cosx')
fig.legend(loc='upper center', frameon=False)

ax[0].set_title('Grafik tanx')
ax[1].set_title('Grafik sinx')
ax[2].set_title('Grafik cosx')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(figsize=(12,8))
ax.plot(x, y)
ax.set_title('Grafik CosX')
cosx5=ax.inset_axes([-0.1,0.9,0.3,0.3])
cosx5.plot(x,y, color='b')
cosx5.plot(x,y5, color='y')
cosx5.set_title('Grafik CosX^5')
cosx2=ax.inset_axes([-0.1,-0.1,0.3,0.3])
cosx2.plot(x,y, color='b')
cosx2.plot(x,y2, color='r')
cosx2.set_title('Grafik CosX^2')
cosx4=ax.inset_axes([0.9,0.9,0.3,0.3])
cosx4.plot(x,y, color='b')
cosx4.plot(x,y4, color='black')
cosx4.set_title('Grafik CosX^4')
cosx3=ax.inset_axes([0.9,-0.1,0.3,0.3])
cosx3.plot(x,y, color='b')
cosx3.plot(x,y3, color='g')
cosx3.set_title('Grafik CosX^3')
plt.show()
```
