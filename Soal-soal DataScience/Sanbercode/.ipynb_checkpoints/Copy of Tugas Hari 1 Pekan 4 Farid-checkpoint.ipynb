<h1>Soal 2: Pemahaman Machine Learning</h1>

Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian aplikasikan untuk membuat model bagi data (variable x, y) di bawah ini. Kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

```python
import matplotlib.pyplot as plt
import numpy as np
 
rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)
 
# data yang akan di prediksi
x_new = np.arange(20, 30, 0.5)
 
fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y)
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.show()
```

```python
# import the regressor 
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
from sklearn.ensemble import RandomForestRegressor
# create a regressor object 
regressor = DecisionTreeRegressor(max_depth=4,random_state = 3,
                                  #min_samples_split=2,
                                  min_samples_leaf=5,
                                  max_leaf_nodes=7)
                                  #max_features=2,  
#criterion='mse', splitter='best', max_depth=None, min_samples_split=2, min_samples_leaf=1, 
#min_weight_fraction_leaf=0.0, max_features=None, random_state=None, max_leaf_nodes=None, 
#min_impurity_decrease=0.0, min_impurity_split=None, presort='deprecated', ccp_alpha=0.0
  
# fit the regressor with X and Y data 
X=x[:,np.newaxis]
X_new=x_new[:,np.newaxis]
Y=y[:,np.newaxis]
regressor.fit(X, Y) 
y_pred = regressor.predict(X)
y_pred2 = regressor.predict(X_new)
fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x,y,label='Data Asli')
ax.scatter(x, y_pred,color='y',label='Data Prediksi')
ax.plot(X_new, y_pred2, color='g',label='Data Baru')
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
 
 
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, random_state = 3,
                           max_depth=4,
                          #min_samples_split=2,
                           min_samples_leaf=5,
                           max_leaf_nodes=17)
# Train the model on training data
#rf.fit(X, Y);
#y_pred_rf = rf.predict(X)
 
#ax.scatter(x, y_pred_rf, color='r',label='Random Forest')
fig.legend()
plt.show()
 
#import sklearn.datasets as datasets
plt.figure(figsize=(25,10))
a = plot_tree(regressor, 
              #feature_names=X, 
              #class_names=y_pred, 
              filled=True, 
              rounded=True, 
              fontsize=18)
```
